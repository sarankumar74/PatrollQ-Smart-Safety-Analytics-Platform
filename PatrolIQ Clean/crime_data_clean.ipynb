```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"C:/Users/saran/OneDrive/Desktop/PatrollQ/PatrolIQ Datasets/chicago_crime_main_data_fast.csv"

df = pd.read_csv(path)
df.head(2)
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce', utc=False)
```

```python
print(df['date'].dtype)
print(df['date'].head())
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['time'] = df['date'].dt.time
df['hour'] = df['date'].dt.hour
df['month_name'] = df['date'].dt.month_name()
```

```python
print("Shape of data:", df.shape)         
print("\nInfo about data:")
print(df.info())                     
```

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
df["ward"] = df["ward"].astype("Int16")
df["community_area"] = df["community_area"].astype("Int16")
df["beat"] = df["beat"].astype("Int32")
```

```python
df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")
df["updated_on"] = pd.to_datetime(df["updated_on"], errors="coerce")
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df['time'].dtypes
```

```python
df['time'].value_counts()
```

```python
df['time_float'] = (
    df['time'].apply(lambda x: x.hour * 3600 + x.minute * 60 + x.second) / 3600
)
```

```python
df['time_float']
```

```python
print("\nData types before:")
print(df.dtypes)
```

```python
df['ward'].iloc[1:5]
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df["ward"].fillna(df["ward"].median(), inplace=True)
df["community_area"].fillna(df["community_area"].median(), inplace=True)
df["x_coordinate"].fillna(df["x_coordinate"].median(), inplace=True)
df["y_coordinate"].fillna(df["y_coordinate"].mean(), inplace=True)
df["latitude"].fillna(df["latitude"].median(), inplace=True)
df["longitude"].fillna(df["longitude"].mean(), inplace=True)
```

```python
for col in ["location_description", "location"]:
    df[col].fillna(df[col].mode()[0], inplace=True)
```
